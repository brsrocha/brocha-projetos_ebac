## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
# importando bibliotecas
import pandas as pd
import numpy as np

# gerando a quantidade de acertos pra cada aluno com base num índice de 0 a 100.
np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])
df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])
df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])
df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])
df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])
df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

Primeiramente, para começarmos com essa atividade, iremos criar um DataFrame concatenado entre todos os DF anteriores, assim, teremos um único DataFrame mostrando as notas de todos os alunos, e poderemos operar em cima deste DataFrame único.

````python
# criando um novo DataFrame, usando os DataFrames atribuídos para cada nota.
df = pd.DataFrame({
    'Matemática': df_mat['Qt_acertos'].values,  
    'Português': df_por['Qt_acertos'].values,
    'Geografia': df_geo['Qt_acertos'].values,
    'Inglês': df_ing['Qt_acertos'].values,
    'História': df_his['Qt_acertos'].values,
    'Física': df_fis['Qt_acertos'].values,
    'Química': df_qui['Qt_acertos'].values
})


df
````

````python
# 1) Calculando a taxa de acertos pra cada aluno
# Vamos criar um novo DataFrame, para que comporte a taxa de acertos de cada um dos alunos nos índices com base na qtd. de questões de cada matéria

percentuais = pd.DataFrame({
    'Matemática': (df['Matemática'] / 24) * 100,
    'Português': (df['Português'] / 18) * 100,
    'Geografia': (df['Geografia'] / 8) * 100,
    'Inglês': (df['Inglês'] / 8) * 100,
    'História': (df['História'] / 8) * 100,
    'Física': (df['Física'] / 12) * 100,
    'Química': (df['Química'] / 12) * 100
}).round(2)

percentuais
````

````python
# 2) Total de Acertos
# vamos criar uma nova coluna no DataFrame, que comporta o número total de acertos de cada aluno

df['Total de Acertos'] = df.sum(axis=1)
df
````

````python
# 3) Calculando o percentual geral de cada aluno, considerando que o total de questões é 90:

df['Percentual Geral'] = ((df['Total de Acertos'] / 90) * 100).round(2)

df
````

````python
# 4) Aplicando uma nota de corte de '45' para os aprovados
# vamos criar uma nova coluna que filtra apenas os valores acima de 45, para sabermos quantos alunos seriam aprovados nesse vestibular com nota de corte de 45.

df['Aprovado'] = df['Percentual Geral'] > 45

aprovados = df[df['Aprovado'] == True]
aprovados
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# 1) Criando um novo DataFrame com os alunos desqualificados

df_desqualificados = df[
    (df['Matemática'] == 0) | 
    (df['Física'] == 0) | 
    (df['Química'] == 0)
]

df_desqualificados
````

````python
# 2) Quantidade de alunos desqualificados

total_desqualificados = len(df_desqualificados)
total_desqualificados
````

````python
# 3) Média dos alunos desqualificados em História e Geografia:

media_historia = (((df_desqualificados['História'].mean()) / 8) * 100).round(2)
media_geografia = (((df_desqualificados['Geografia'].mean()) / 8) * 100).round(2)

print(media_historia, media_geografia)
````

````python
# 4) total de alunos que foram aprovados, ou seja, que não tiveram nota menor de 45 e nem foram desqualificados:

df_aprovados = df[
    (df['Aprovado'] == True) &
    (df['Matemática'] != 0) &
    (df['Física'] != 0) &
    (df['Química'] != 0)
]

df_aprovados
````

Existe uma forma ainda mais curta de criar o df de aprovados, usando a função `.index` e `.isin()`, uma função para extrair o index e uma função para checar se um determinado valor está em um determinado local, respectivamente:

````python
df_aprovados_novo = df[
    (~df.index.isin(df_desqualificados.index)) &  # NÃO (~) está no index de desqualificados
    (df['Percentual Geral'] > 45) # Tem nota maior que 45
]

df_aprovados_novo
````

## 4) Vacinação II
Gere um *DataFrame* que contenha somente os estabelecimentos que aplicaram vcinas a menores de 18 anos. Nesse *DataFrame* devem conter somente os dados dos estabelecimentos, mais uma coluna sendo a quantidade de vacinas que o estabelecimento aplicou a menores de 18 anos.  
  
1. crie uma cópia do *DataFrame* original, contendo somente os registros de vacinas realizadas a menores de 18 anos.  
2. crie uma lista das colunas desse *DataFrame* com o atributo de *DataFrame* **.columns()**  
3. Nesse *DataFrame* faça uma contagem do campo ```vacina_categoria_nome```.
3. a partir da lista de colunas, escolha somente aquelas que são referentes ao estabelecimento, faça uma lista com esses valores.  
4. usando o método *.loc*, selecione somente essas variáveis  
5. Aplique o método **.drop_duplicates** e crie uma lista com uma linha para cada estabelecimento, com os dados do estabelecimento  

````python
# 1) Criando um novo DataFrame apenas com os dados dos vacinados com menos de 18 anos:

df_menor_18 = df_vacinas[df_vacinas['paciente_idade'] < 18]
df_menor_18
````

````python
# 2) Criando uma lista com as colunas do DataFrame

lista_colunas = df_menor_18.columns.tolist()
lista_colunas
````

No exercício 3, ficou meio ambíguo o que é pra ser feito. Inseri duas tentativas distintas para resolução do exercício, uma que conta a quantidade de valroes únicos na coluna 'vacina_categoria_nome', e a outra que conta a quantidade de vacinações de cada estabelecimento:

````python
# 3, i) Contando valores únicos de vacina_categoria_nome no DataFrame:

df_menor_18['vacina_categoria_nome'].nunique()
````

````python
# 3, ii) Contando a quantidade de vacinas de cada estabelecimento:

contagem_vacinas = df_menor_18.groupby('estabelecimento_razaoSocial').size().reset_index(name='quantidade_vacinas').sort_values(by='quantidade_vacinas', ascending=False)
contagem_vacinas
````

O método abaixo uiliza a lista com as colunas do DataFrame que utilizamos anteriormente, e usa um recurso chamado *list comprehension* do Python para filtrar apenas os elementos da lista que possuem 'estabelecimento' nelas. Esse recurso pode ser melhor entendido aqui: https://www.w3schools.com/python/python_lists_comprehension.asp
Uma list comprehension facilita o processo de filtragem e manipulação de listas. Como nós já criamos uma lista com as colunas do DataFrame, usamos list comprehension para facilitar o processo de achar todas os itens contendo 'estabelecimento'.

````python
# 4) Selecionando a lista de estabelecimentos usando list comprehension do Python:

colunas_estabelecimento = [col for col in lista_colunas if 'estabelecimento' in col] # esse código checa todas as colunas do DF e armazena em 'colunas_estabelecimento' todos que retornam True para 'estabelecimento' in col
colunas_estabelecimento
````

````python
df_menor_18.columns
````

Como já temos uma lista de estabelecimentos, podemos usar essa lista para filtrar diretamente o DataFrame:

````python
# 5) Usando a lista de estabelecimentos para filtrar o DataFrame:

df_estabelecimentos = df_menor_18.loc[:, colunas_estabelecimento]
df_estabelecimentos
````

````python
# 6) Por fim, iremos remover os duplicados do DataFrame anterior que criamos, assim obtendo um DataFrame com todos os dados de estabelecimento únicos:

df_estabelecimentos = df_estabelecimentos.drop_duplicates()
df_estabelecimentos
````
