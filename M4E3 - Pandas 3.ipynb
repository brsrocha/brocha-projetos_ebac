### Bônus: Corrigindo problema com as URLs de API do governo e melhorando a função:
Neste exercício, usamos uma API do governo que nos fornece dados em `json` de solicitações de PVL para Estados e Municípios brasileiros. Contudo, o site atualizou desde quando foi utilizado nas aulas do curso, e o link que era utilizado não funciona mais de forma direta, parece que o site quebrou e o governo migrou para uma nova plataforma em (https://sadipem.tesouro.gov.br/sadipem/private/pages/pvl/pvl_list.jsf), que exporta arquivos zipados ao invés de fornecer um link direto para o CSV.
Por isso, para resolver as questões da atividade acima, já que não tinha o site para extrair os dados com o filtro, eu tive que analisar se as URLs tinham algum padrão que eu poderia mudar algum elemento para chamar algum dado específico, e de fato isto ocorria, existe dois identificadores nas URL, que ao alterar para a sigla do estado e para o tipo_interessado, podemos obter a URL de qualquer solicitação, sem precisar do site:

`uf={**SIGLA DO ESTADO**}`
`tipo_interessado={**TIPO (Estado, Município)**}`

Assim, por exemplo, apenas alterando essa UF para a sigla do estado que queremos extrair os dados, já podemos puxar os dados para o estado ou para o município especificado, sem necessidade de entrar no site do governo pra isso. Por exemplo, para acessar os dados do estado e dos municípios de São Paulo, é apenas uma mudança no URL:

*   **Estado de SP:** https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Estado
*   **Municípios de SP:** https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Munic%C3%ADpio

Note como, nos exemplos acima, as únicas coisas que mudam de um estado para o outro são os identificadores ``UF = {sigla}``, e  ``tipo_interessado = {tipo}``

Com base nisso, decidi escrever uma nova função. Esta nova função, ao invés de requisitar que o usuário pré-defina uma variável contendo a URL dos dados que ele quer acessar, para depois chamar a função (como fizemos com a função 'interessado' anterior), criei uma função mais flexível, que permite que o usuário use como argumento apenas a *sigla do Estado* que ele quer acessar, e o *tipo interessado*, sem necessidade de definir nenhuma variável para seu uso. Confira a função abaixo, com comentários explicando o código:

````python
# Criando a função, que têm como argumento o UF e o tipo interessado, que pode ser 'Estado' ou 'Município':
def pvl_estado(UF, tipo_interessado):
    url_estado_1 = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf='        # Vamos pegar a primeira parte do URL, e separar do identificador de UF=
    url_estado_2 = '&tipo_interessado='                                                # Vamos então pegar a última parte do URL, e também separadr do identificador tipo_interessado
    lista_uf = ['AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN', 'RS', 'RO', 'RR', 'SC', 'SP', 'SE', 'TO']
    lista_tipos = ['Estado', 'Município']

    if UF in lista_uf:                                                                 # usando if statement para evitar parâmetros inválidos de UF e tipo
        if tipo_interessado in lista_tipos:
            url_nova = f'{url_estado_1}{UF}{url_estado_2}{tipo_interessado.capitalize()}'
            r_estado = requests.get(url_nova)
            status_request = r_estado.status_code
            dados_estado = r_estado.json()
        else:
            return "Tipo Interessado inválido, por favor, escolha 'Estado' ou 'Município' "
        
        return pd.DataFrame(dados_estado['items'])
    
    else:
        return "UF inválida, por favor insira uma UF válida."
````

Testando com dados válidos:

````python
dados_municipio_para = pvl_estado(UF="PA", tipo_interessado="Município")
dados_municipio_para
````
